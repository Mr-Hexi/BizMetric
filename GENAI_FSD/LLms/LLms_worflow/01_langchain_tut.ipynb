# Basic LangChain Chatbot

````python
from dotenv import load_dotenv
import os
from langchain_google_genai import ChatGoogleGenerativeAI
````

````python

load_dotenv()  # Load environment variables from .env file

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

response = llm.invoke("Explain what revenue growth means.")

print(response.content)
````

## Prompt Templates

Prompt templates make prompts reusable and dynamic.

````python
from langchain_core.prompts import ChatPromptTemplate
````

````python
prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple terms."
)
````

````python
chain = prompt | llm
````

````python
response = chain.invoke({"topic": "financial forecasting"})
````

````python
print(response.content)
````

## Tool Calling

Now the chatbot can use external functions.

````python
from langchain_core.tools import tool
````

````python
@tool
def add_numbers(a: int, b: int) -> int:
    """Add two numbers together."""
    return a + b

````

````python
from langchain.agents import create_agent
agent = create_agent(
    model=llm,
    tools=[add_numbers],
    system_prompt="You are a helpful assistant."
)
````

Why messages Format Is Used

Modern chat models don't work with a single string.

They work with conversation messages.

- role: user
- role: assistant
- role: system

system: You are a math assistant

user: What is 5 + 3?

assistant: 8

This allows the agent to keep track of:

- conversation history
- tool calls
- intermediate reasoning


```
User question
 ↓
LLM decides to call tool
 ↓
Tool returns result
 ↓
LLM generates final answer
```

````python
result = agent.invoke({
    'messages':[
        {"role": "user", "content": "What is 5 + 3?"}
    ]
})
````

````python
print(result)
````

````python
print(result['messages'][-1].content)
````
