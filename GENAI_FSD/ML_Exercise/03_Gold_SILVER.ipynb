```python
import yfinance as yf
import pandas as pd

# Gold and Silver tickers
tickers = ["GC=F", "SI=F"]

# Download last 1 year data
data = yf.download(
    tickers,
    period="1y",
    interval="1d"
)

# Keep closing prices
prices = data["Close"]

# Rename columns
prices.columns = ["Gold", "Silver"]

print(prices.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
prices.plot()

plt.title("Gold vs Silver Prices (Past 1 Year)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
correlation = prices.corr()

print(correlation)
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm"
)

plt.title("Gold vs Silver Correlation")
plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.scatter(prices["Gold"], prices["Silver"])

plt.xlabel("Gold Price")
plt.ylabel("Silver Price")
plt.title("Gold vs Silver Correlation Scatter")

plt.show()
```

```python
st_date = '2020-01-01'
today = datetime.date.today()
```
